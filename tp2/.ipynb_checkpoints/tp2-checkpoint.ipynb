### Linear Method

We first try to perform camera calibration using the linear method. From the lecture, we know that it ideally solves an equation ($\mathbf 0$ being a vector, and $\mathbf p$ a vector representation of the camera matrix)

$$A \mathbf p = \mathbf 0$$

**1.** Build the matrix `A` that defines this calibration equation. What are its dimensions?

```python
#Q1
xdim = np.shape(pts2d)[0]
A = np.zeros((2*xdim, 12))
print("Dimensions of A are: (2*{}, 12) = {}".format(xdim, A.shape))

#axi = (0,0,0,0, -Xi, -Yi, -Zi, -1, yiXi, yiXi, yiZi, yi)^T
#ayi = (-Xi, -Yi, -Zi, -1,0,0,0,0,xiXi, xiYi, xiZi, xi) ^ T
# Ai = (axi, ayi)^T 

for i in range(xdim):
    A[2*i, 4:7] = -pts3d[i, :]
    A[2*i, 7] = -1
    A[2*i, 8:11] = pts2d[i, 1]*pts3d[i, :] 
    A[2*i, 11] = pts2d[i, 1]
    A[2*i+1, :3] = -pts3d[i, :]
    A[2*i+1, 3] = -1
    A[2*i+1, 8:11] = pts2d[i, 0]*pts3d[i, :] 
    A[2*i+1, 11] = pts2d[i, 0]
```

**2.a)** Perform SVD (using `np.linalg.svd`) on `A` <br/>
**b)** Check that the SVD is coherent and that one of the singular value is close to $0$ <br/>
**c)** Use this SVD to compute the camera matrix `P`.

```python
# Q2.a
U,s,V = np.linalg.svd(A)
S = np.zeros(np.shape(A))
np.fill_diagonal(S, s)

# Q2.b
print("Dimensions: ", np.shape(U),np.shape(S), np.shape(V))
print("USV =? A:\n", np.isclose(np.dot(U, np.dot(S, V)),A))

#Q2.c
i = np.where(s == np.min(s))[0]
P = V[i,:].reshape(3, 4)
```

**3.** Write a function `projection_error(P, pts3d, pts2d)` that computes the measurement error (Mean Squared Error) between the 3D points projected with a matrix `P`, and the 2D points.

```python
# Q3 

def projection_error(P, pts3d=pts3d, pts2d=pts2d):
    # euclidean in 2D/3D
    n = np.shape(pts3d)[0]
    p2d = np.c_[pts2d, np.ones(n)]
    p3d = np.c_[pts3d, np.ones(n)]
    # sum ((xi - P1.Xi/ P3.Xi)**2 + (yi - P2.Xi/ P3.Xi)**2)
    res = 0
    for i in range(n):
        a = pts2d[i, 0] - np.dot(P[0], p3d[i, :])/ np.dot(P[2], p3d[i, :])
        b = pts2d[i, 1] - np.dot(P[1], p3d[i, :])/ np.dot(P[2], p3d[i, :])
        res += a**2 + b**2
    return res

err = projection_error(P)
print("Error = ", err)
```

**4.** Visualize the projected 3D points and the 2D points on the same figure (use `plt.scatter` with different `marker` options)

```python
# Q4
#projection of 3d points
P3d = np.c_[pts3d, np.ones(xdim)]
proj = np.dot(P3d, P.T)
lx, ly = [],[]
for p in proj: lx.append(p[0]/p[2]); ly.append(p[1]/p[2])

#plot
plt.figure(figsize = (10,5))
plt.scatter(pts2d[:,0], pts2d[:, 1], marker = '8' ,s = 80, c='red', label = "pts2d")
plt.scatter(lx,ly, marker = '*', c = 'yellow', label="projected 3d")
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title("2D points and the Projected 3D points")
plt.show()
```

### Camera parameters

We want to extract the parameters of the camera from the matrix `P`, *i.e.* the internal $(3\times 3)$ calibration matrix $\mathcal K$, the external $(3\times 3)$ rotation matrix $R$ and the $(3\times 1)$ translation vector $\mathbf t$ so that:

$$P=\mathcal K [R | \mathbf t]$$

To do so, we are going to follow the formulas from the lectures (starting around slide 73), and the numpy functions (`np.linalg.norm`, `np.linalg.inv`, `np.dot`, `np.cross`...)

**5.a)** Write a function `camera_parameters(P, eps=1)` that takes as input the camera matrix `P` and a sign `eps` ($\varepsilon$, the sign of $\rho$ in the equations from the lecture), and outputs three values `K`, `R` and `t` that correspond to the intrinsic and extrinsic parameters of the camera. <br/>
**b)** Display the results for `eps=1` and check that your values `(K, R, t)` can correctly reconstruct `P`.

```python
# Q5.a
def camera_parameters(P, eps=1):
    a1, a2, a3, b  = P[0,:3], P[1,:3], P[2,:3], P[:,-1]
    rho = eps / np.linalg.norm(a3)
    a13, a23 = np.cross(a1,a3), np.cross(a2,a3)
    
    #R
    r3 = rho * a3
    r1 = 1/np.linalg.norm(a23)* np.cross(a2, a3)
    r2 = np.cross(r3, r1)
    R = np.array([r1,r2,r3])
    
    # K
    u0, v0 = rho**2 * a1@a3, rho**2 * a2@a3
    costh = - a13@a23/(np.linalg.norm(a13) * np.linalg.norm(a23))
    #sinth = - np.sin(np.arccos(costh)) 
    sinth = -a13@a2/(np.linalg.norm(a13) * np.linalg.norm(a23)) * np.linalg.norm(a3)
    alpha = rho**2 * np.linalg.norm(a13)*sinth
    beta =  rho**2 * np.linalg.norm(a23)*sinth
    coth = costh / sinth
    K = np.array([[alpha, -alpha*coth, u0], [0, beta/sinth, v0], [0, 0, 1]])
    
    #t
    t = rho * np.linalg.inv(K) @ b
    return K, R, t, rho

K, R, t, rho = camera_parameters(P)

# Q3.b 
print('K={}\n\n R={} \n\n t={}\n\n'.format(K,R,t))
KRt = K.dot(np.c_[R,t])
print('pP =? KRt:\n', np.isclose(rho*P, KRt))
```

**6.** The sign of $\rho$ is not easy to decide a priori: test your function with both choices of `eps`, and comment on which option makes sense (is the object in front of the camera?)

```python
# Q6
K2, R2, t2, rho2 = camera_parameters(P,-1) 
print('K={}\n\n R={} \n\n t={}\n\n'.format(K2,R2,t2))
KRt2 = K2.dot(np.c_[R2,t2])
print('pP =? KRt2:\n', np.isclose(rho2*P, KRt2))
```

**Comment**: For this case the sign of $\rho$ should be positive since re-construction of $P$ is not done correctly with $\varepsilon = -1$. This means that the object is behind the camera.

### Non linear method

**7.a)** Improve the estimate of the camera matrix using non linear least square: use `projection_error` from **3.** and `optimize.leastsq` which implements a variant of Levenberg-Marquardt (read the documentation to use it in a sensible way). <br/>
**b)** Compare the reprojection error before and after the optimization.

```python
# Q7.a
def projection_error2(P, pts3d=pts3d, pts2d=pts2d):
    # euclidean in 2D/3D
    P = P.reshape(3,4)
    n = np.shape(pts3d)[0]
    p2d = np.c_[pts2d, np.ones(n)]
    p3d = np.c_[pts3d, np.ones(n)]
    # sum ((xi - P1.Xi/ P3.Xi)**2 + (yi - P2.Xi/ P3.Xi)**2)
    res = np.zeros((2*n))
    for i in range(n):
        a = pts2d[i, 0] - np.dot(P[0], p3d[i, :])/ np.dot(P[2], p3d[i, :])
        b = pts2d[i, 1] - np.dot(P[1], p3d[i, :])/ np.dot(P[2], p3d[i, :])
        res[2*i] = a
        res[2*i + 1] = b
    return res

    
Pnew = optimize.leastsq(projection_error2, P.flatten())[0].reshape(3,4)
```

```python
# Q7.b 
print("E_before = {}, E_after = {}".format(projection_error(P), projection_error(Pnew)))
```

**Comment**: Using Levenberg-Marquardt we see a little improvement in the error.
